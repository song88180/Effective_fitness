## Simulations with allele-frequency-dependent selection

```python
# sample s
N = 1000
State_list = []
fitness_gmean_list = []
fitness_e_list = []
drift_list_list = []
Ngen_list = []
N_rep = 2000
N_stop = 30000
rep = 0

#nrand.seed(73)

# Find a simulaiton where fG < 1 significantly.
while rep < N_rep:
    A = 1
    s = -0.5

    dAF_list = []
    fitness_list = []
    dAF_drift_list = []
    dAF_sel_list = []
    AF_list = [A/N]
    for N_gen in range(1,N_stop+1):
        if (A >= 1000-10):
            s = -0.1
        else:
            s = 0.01
        AF = A/N
        dAF = s*AF*(1-AF)/(1+s*AF)
        dAF_sel_list.append(dAF)
        fitness_list.append(1+s)
        P = A*(1+s)/(s*A+N)
        A = nrand.binomial(N,p=P)
        dAF_drift_list.append(A/N-P)
        AF_list.append(A/N)
        if (A == N) or (A == 0):
            break

    if (A == N):
        fitness_gmean = np.product(fitness_list)**(1/len(fitness_list))
        selcoef_list = np.array(fitness_list)-1
        dAF_mean = np.mean(dAF_sel_list)
        fitness_effective = (1+2*dAF_mean)/(1-2*dAF_mean)

        fitness_gmean_list.append(fitness_gmean)
        fitness_e_list.append(fitness_effective)
        drift_list_list.append(np.sum(dAF_drift_list))
        Ngen_list.append(N_gen)
        print(f'fE:{fitness_effective}, fG:{fitness_gmean}')
        if (fitness_effective > 1) and (fitness_gmean < 1):
            print(np.sum(dAF_sel_list)/(np.sum(dAF_sel_list) + np.sum(dAF_drift_list)))
            break
    
    rep += 1
    if rep%100 == 0:
        print(rep,end='\r',flush=True)
```

# Plot an example run

```python
# Prepare variables
Ne=1000
mean_f = fitness_gmean
effective_f = fitness_effective

# Plot allele frequency dynamics
plt.subplots(figsize=[10,4],facecolor='white')
plt.plot(AF_list)

# Plot dAF_sel
for i,dAF_sel in enumerate(dAF_sel_list):
    plt.plot([i,i],[AF_list[i],AF_list[i]+dAF_sel*50],color='r',alpha=0.5,lw=0.5)
    
# Plot dAF_drift
for i,dAF_drift in enumerate(dAF_drift_list):
    plt.plot([i,i],[AF_list[i],AF_list[i]+dAF_drift*10],color='b',alpha=0.5, lw=0.5)
    
# Plot shaded area
plt.fill_between(range(len(AF_list)), y1=-0.1, y2=1.1,
    where= (np.array(AF_list) >= 1-10/(Ne)),
    facecolor='gray', alpha=0.5,linewidth=0)

# Plot texts
plt.text(s='Geometric mean fitness ($\mathit{f}_G$) ' + f'= {mean_f:.4f}',
         x=500,y=0.3,fontsize=15)
plt.text(s=f's = -0.1 in gray\ns = 0.01 in white',
         x=50,y=0.7,fontsize=15)

plt.ylabel('Allele frequency')
plt.xlabel('Generations')

plt.xlim(0,len(AF_list))
plt.ylim(-0.1,1.1)
#plt.savefig('Figure1_1.pdf',pad_inches=0.5,bbox_inches='tight')
```
